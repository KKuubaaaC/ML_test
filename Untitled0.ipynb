```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# cieki do naszych przykadowych plikw
image_path = '/content/drive/MyDrive/Viral Pneumonia/images/Viral Pneumonia-1.png'
mask_path = '/content/drive/MyDrive/Viral Pneumonia/masks/Viral Pneumonia-1.png'
IMG_SIZE = 128 # Definiujemy standardowy rozmiar obrazu

def preprocess_data(img_path, mask_path, size):
    """
    Wczytuje obraz i mask, zmienia ich rozmiar i normalizuje wartoci.
    """
    # Wczytanie obrazu i konwersja do RGB
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (size, size))
    img = img / 255.0 # Normalizacja obrazu do [0, 1]

    # Wczytanie maski w skali szaroci
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    mask = cv2.resize(mask, (size, size))
    mask = mask / 255.0 # Normalizacja maski do [0, 1]

    # Rozszerzenie wymiaru maski, aby miaa ksztat (size, size, 1)
    # Jest to standardowy format dla masek w zadaniach segmentacji
    mask = np.expand_dims(mask, axis=-1)

    return img, mask

# ------ TESTOWANIE NASZEJ FUNKCJI ------

# Uywamy naszej nowej funkcji do przetworzenia danych
processed_image, processed_mask = preprocess_data(image_path, mask_path, IMG_SIZE)

# Wywietlamy wyniki, aby zobaczy efekt
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(processed_image)
plt.title(f'Obraz po przetworzeniu ({IMG_SIZE}x{IMG_SIZE})')
plt.axis('off')

plt.subplot(1, 2, 2)
# Uywamy .squeeze(), aby usun dodatkowy wymiar z maski na potrzeby wizualizacji
plt.imshow(processed_mask.squeeze(), cmap='gray')
plt.title(f'Maska po przetworzeniu ({IMG_SIZE}x{IMG_SIZE})')
plt.axis('off')

plt.show()

# Drukujemy informacje, aby potwierdzi, e wszystko si zgadza
print(f"Ksztat przetworzonego obrazu: {processed_image.shape}")
print(f"Ksztat przetworzonej maski: {processed_mask.shape}")
print(f"Typ danych obrazu: {processed_image.dtype}")
print(f"Typ danych maski: {processed_mask.dtype}")
print(f"Unikalne wartoci w przetworzonej masce: {np.unique(processed_mask)}")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# cieki do naszych przykadowych plikw
image_path = '/content/drive/MyDrive/Viral Pneumonia/images/Viral Pneumonia-1.png'
mask_path = '/content/drive/MyDrive/Viral Pneumonia/masks/Viral Pneumonia-1.png'
IMG_SIZE = 128

def preprocess_data(img_path, mask_path, size):
    """
    Wczytuje obraz i mask, zmienia ich rozmiar i normalizuje wartoci.
    """
    # Przetwarzanie obrazu (bez zmian)
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (size, size))
    img = img / 255.0

    # Przetwarzanie maski (z poprawk)
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    mask = cv2.resize(mask, (size, size))
    mask = mask / 255.0

    # --- NOWA LINIA: USTAWIANIE PROGU ---
    # Wszystkie piksele o wartoci > 0.5 zamie na 1, a reszt na 0.
    mask = (mask > 0.5).astype(np.float32)

    mask = np.expand_dims(mask, axis=-1)

    return img, mask

# ------ TESTOWANIE POPRAWIONEJ FUNKCJI ------

processed_image, processed_mask = preprocess_data(image_path, mask_path, IMG_SIZE)

# Wywietlanie wynikw (bez zmian)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(processed_image)
plt.title(f'Obraz po przetworzeniu ({IMG_SIZE}x{IMG_SIZE})')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(processed_mask.squeeze(), cmap='gray')
plt.title(f'Maska po przetworzeniu ({IMG_SIZE}x{IMG_SIZE})')
plt.axis('off')
plt.show()

# Sprawdmy, czy teraz wartoci s poprawne
print(f"Ksztat przetworzonej maski: {processed_mask.shape}")
print(f"Typ danych maski: {processed_mask.dtype}")
print(f"Unikalne wartoci w przetworzonej masce: {np.unique(processed_mask)}")
```

```python
import os
from sklearn.model_selection import train_test_split

# Definiujemy gwne cieki
base_dir = '/content/drive/MyDrive/Viral Pneumonia/'
images_dir = os.path.join(base_dir, 'images')
masks_dir = os.path.join(base_dir, 'masks')

# Pobieramy posortowane listy plikw z obu folderw
# Sortowanie jest kluczowe, aby mie pewno, e image_0 odpowiada masce_0 itd.
image_files = sorted(os.listdir(images_dir))
mask_files = sorted(os.listdir(masks_dir))

# Tworzymy pene cieki do kadego pliku
image_paths = [os.path.join(images_dir, fname) for fname in image_files]
mask_paths = [os.path.join(masks_dir, fname) for fname in mask_files]

# Sprawdzamy, czy liczba obrazw i masek si zgadza
print(f"Znaleziono {len(image_paths)} obrazw.")
print(f"Znaleziono {len(mask_paths)} masek.")

if len(image_paths) != len(mask_paths):
    print("Bd: Liczba obrazw i masek nie jest taka sama!")
else:
    # Dzielimy dane na zbir treningowy (80%) i walidacyjny (20%)
    # `random_state=42` zapewnia, e podzia bdzie zawsze taki sam przy ponownym uruchomieniu
    train_images, val_images, train_masks, val_masks = train_test_split(
        image_paths, mask_paths, test_size=0.2, random_state=42
    )

    print("\nPodzia danych zakoczony sukcesem!")
    print(f"Liczba prbek w zbiorze treningowym: {len(train_images)}")
    print(f"Liczba prbek w zbiorze walidacyjnym: {len(val_images)}")

    # Wywietlmy przykadow par ze zbioru treningowego, aby si upewni
    print("\nPrzykadowa para ze zbioru treningowego:")
    print(f"Obraz: {train_images[0]}")
    print(f"Maska: {train_masks[0]}")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Dropout, Conv2DTranspose, concatenate
from tensorflow.keras.models import Model

# Upewnijmy si, e nasza funkcja do preprocessingu jest dostpna
# (jeli uruchamiasz to w nowej komrce, skopiuj j ponownie)
import cv2
IMG_SIZE = 128
def preprocess_data(img_path, mask_path, size):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (size, size))
    img = img / 255.0
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    mask = cv2.resize(mask, (size, size))
    mask = mask / 255.0
    mask = (mask > 0.5).astype(np.float32)
    mask = np.expand_dims(mask, axis=-1)
    return img, mask

# 1. DEFINICJA GENERATORA DANYCH
class DataGenerator(tf.keras.utils.Sequence):
    def __init__(self, image_paths, mask_paths, batch_size=16, dim=(128,128), n_channels=3, shuffle=True):
        self.dim = dim
        self.batch_size = batch_size
        self.image_paths = image_paths
        self.mask_paths = mask_paths
        self.n_channels = n_channels
        self.shuffle = shuffle
        self.on_epoch_end()

    def __len__(self):
        return int(np.floor(len(self.image_paths) / self.batch_size))

    def __getitem__(self, index):
        indexes = self.indexes[index*self.batch_size:(index+1)*self.batch_size]
        batch_image_paths = [self.image_paths[k] for k in indexes]
        batch_mask_paths = [self.mask_paths[k] for k in indexes]
        X, y = self.__data_generation(batch_image_paths, batch_mask_paths)
        return X, y

    def on_epoch_end(self):
        self.indexes = np.arange(len(self.image_paths))
        if self.shuffle == True:
            np.random.shuffle(self.indexes)

    def __data_generation(self, batch_image_paths, batch_mask_paths):
        X = np.empty((self.batch_size, *self.dim, self.n_channels))
        y = np.empty((self.batch_size, *self.dim, 1))

        for i, (img_path, mask_path) in enumerate(zip(batch_image_paths, batch_mask_paths)):
            img, mask = preprocess_data(img_path, mask_path, self.dim[0])
            X[i,] = img
            y[i,] = mask
        return X, y

# 2. DEFINICJA MODELU U-NET
def build_unet(input_shape):
    inputs = Input(input_shape)

    # cieka enkodera (kurczca si)
    c1 = Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(inputs)
    c1 = Dropout(0.1)(c1)
    c1 = Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c1)
    p1 = MaxPooling2D((2, 2))(c1)

    c2 = Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p1)
    c2 = Dropout(0.1)(c2)
    c2 = Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c2)
    p2 = MaxPooling2D((2, 2))(c2)

    # Most
    c5 = Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p2)
    c5 = Dropout(0.2)(c5)
    c5 = Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c5)

    # cieka dekodera (rozszerzajca si)
    u6 = Conv2DTranspose(32, (2, 2), strides=(2, 2), padding='same')(c5)
    u6 = concatenate([u6, c2])
    c6 = Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u6)
    c6 = Dropout(0.1)(c6)
    c6 = Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c6)

    u7 = Conv2DTranspose(16, (2, 2), strides=(2, 2), padding='same')(c6)
    u7 = concatenate([u7, c1])
    c7 = Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u7)
    c7 = Dropout(0.1)(c7)
    c7 = Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c7)

    # Warstwa wyjciowa
    outputs = Conv2D(1, (1, 1), activation='sigmoid')(c7)

    model = Model(inputs=[inputs], outputs=[outputs])
    return model

# 3. Tworzymy instancj modelu i wywietlamy jego podsumowanie
input_shape = (IMG_SIZE, IMG_SIZE, 3)
model = build_unet(input_shape)
model.summary()

# 4. Tworzymy instancje generatorw
BATCH_SIZE = 16
train_generator = DataGenerator(train_images, train_masks, batch_size=BATCH_SIZE, dim=(IMG_SIZE, IMG_SIZE))
val_generator = DataGenerator(val_images, val_masks, batch_size=BATCH_SIZE, dim=(IMG_SIZE, IMG_SIZE), shuffle=False)

print("\nModel i generatory danych zostay pomylnie zdefiniowane.")
```

```python
# 1. Kompilacja modelu
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 2. Trening modelu
# Trenujemy na razie tylko przez 15 epok, aby zobaczy, jak model si uczy.
# W przyszoci mona zwikszy t liczb.
EPOCHS = 15
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=EPOCHS
)

# 3. Wizualizacja wynikw treningu
plt.figure(figsize=(12, 5))

# Wykres dokadnoci
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Dokadno treningowa')
plt.plot(history.history['val_accuracy'], label='Dokadno walidacyjna')
plt.title('Dokadno modelu')
plt.xlabel('Epoka')
plt.ylabel('Dokadno')
plt.legend()

# Wykres straty
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Strata treningowa')
plt.plot(history.history['val_loss'], label='Strata walidacyjna')
plt.title('Strata modelu')
plt.xlabel('Epoka')
plt.ylabel('Strata')
plt.legend()

plt.show()

# 4. Predykcje i wizualna ocena
# Wemy jedn paczk danych z generatora walidacyjnego
val_images_batch, val_masks_batch = val_generator[0]

# Wykonajmy predykcje na tej paczce
predictions = model.predict(val_images_batch)

# Wywietlmy kilka przykadw: oryginalny obraz, prawdziw mask i predykcj modelu
plt.figure(figsize=(10, 20))
num_examples = 5 # Liczba przykadw do wywietlenia
for i in range(num_examples):
    # Oryginalny obraz
    plt.subplot(num_examples, 3, i*3 + 1)
    plt.imshow(val_images_batch[i])
    plt.title("Oryginalny obraz")
    plt.axis('off')

    # Prawdziwa maska
    plt.subplot(num_examples, 3, i*3 + 2)
    plt.imshow(val_masks_batch[i].squeeze(), cmap='gray')
    plt.title("Prawdziwa maska")
    plt.axis('off')

    # Maska przewidziana przez model
    # Ustawiamy prg 0.5 na predykcji, aby uzyska binarn mask
    predicted_mask = (predictions[i] > 0.5).astype(np.uint8)
    plt.subplot(num_examples, 3, i*3 + 3)
    plt.imshow(predicted_mask.squeeze(), cmap='gray')
    plt.title("Przewidziana maska")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
